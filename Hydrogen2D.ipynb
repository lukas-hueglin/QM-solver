# Hydrogen atom in 2D Example

```python
import scipy
import numpy as np
import matplotlib.pyplot as plt

from qmsolver.System2D import System2D
```

```python
%matplotlib widget
```

```python
NUM = 50
WIDTH = 10
```

```python
# create a grid
space = np.mgrid[-WIDTH/2:WIDTH/2:NUM*1j, -WIDTH/2:WIDTH/2:NUM*1j].T
r = np.linalg.norm(space, axis=2)

# create potential
V = 10*np.clip(-1/r, -1000, 1)

# create system
s = System2D(potential=V, width=WIDTH, num=NUM)

# plot potential
fig, ax = plt.subplots()
s.plot_potential(ax=ax)
```

```python
fig, ax = plt.subplots()
s.plot_eigenvector('H', num=16, which='r', ax=ax)
```

```python
fig, ax = plt.subplots()

s.plot_eigenvalues('H', eigval_range=(0, 50), ax=ax)
```
